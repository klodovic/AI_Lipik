```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('Life_Expectancy ~ C(Country) + C(Diet_Type) + C(Country):C(Diet_Type)', data=df).fit()
```

```python
anova_table = sm.stats.anova_lm(model, typ=2)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
angazman = pd.read_csv("angazman.csv")
```

```python
angazman
```

```python
angazman.corr()
```

```python
### Prvi korak bez kojeg smo i mogli, ali ga je bolje napraviti...
```

```python
# PCA je osjetljiv na skalu varijabli, pa ih moramo standardizirati.
# Stvaramo novi DataFrame samo s nezavisnim varijablama za PCA.
X = angazman[['Broj_Posjeta_Web_Stranici_u_Tjednu',
        'Prosjecno_Vrijeme_na_Stranici_Min',
        'Broj_Pregledanih_Proizvoda']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
pca = PCA(n_components=1)
principal_components = pca.fit_transform(X_scaled)
```

```python
angazman["angazman"] = principal_components
```
