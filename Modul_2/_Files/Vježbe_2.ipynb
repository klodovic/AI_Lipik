### Vjeba Bike

```python
bike = pd.read_csv("bike_buyers_clean.csv")
```

```python
bike.head()
```

```python
bike.describe()
```

```python
plt.xlabel('Income')
plt.ylabel('Children')
plt.scatter(bike.Income, bike.Children, color='blue', marker='+')
```

```python
plt.xlabel('Children')
plt.ylabel('Cars')
plt.scatter(bike.Children, bike.Cars, color='blue', marker='+')
```

```python
plt.xlabel("Broj djece")
plt.ylabel("Broj kupaca")
plt.title("Distribucija prema broju djece")
plt.hist (bike.Children, bins = 20)
```

```python
bike['Marital Status'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
bike['Education'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
bike['Region'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
bike['Children'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
bike['Occupation'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
sns.barplot(data=bike, x='Cars', y='Income', hue='Gender')
plt.show()
```

```python

bike.boxplot(column='Age', by='Purchased Bike')
plt.xlabel('Purchased Bike')
plt.ylabel('Age')
plt.title('Age Distribution by Bike Purchase')
plt.suptitle('')
```

```python
plt.boxplot(bike["Income"])
```

```python
sns.violinplot(x='Marital Status', y='Income', data=bike)
plt.title('Income Distribution by Marital Status')
```

```python
corr = bike.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
```
