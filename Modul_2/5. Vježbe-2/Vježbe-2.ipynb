```python
cetvrti = df['cetvrt'].value_counts()
cetvrti
```

```python
cetvrti.plot(kind='bar', color='blue')
```

```python
cetvrti_kupaona = df.groupby(['cetvrt', 'kupaona']).size().unstack(fill_value=0)
cetvrti_kupaona
```

```python
cetvrti_kupaona.plot(kind='bar', color='lightcoral')
```

```python
cetvrti_kupaona.plot(kind='bar', ax=plt.gca(), cmap='viridis')
```

```python
cetvrti_balkon = df.groupby(['cetvrt', 'posjeduje balkon']).size().unstack(fill_value=0)
cetvrti_balkon
```
