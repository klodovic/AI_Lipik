```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as ss
np.random.randint(1,7)
```

### 1. Bacanje kockice

```python
nekoliko_puta = np.random.randint(1,7, size=10)
#print(nekoliko_puta)
puno_puta = np.random.randint(1,7, size=100000)
#print(puno_puta)

malo_bacanja = np.histogram(nekoliko_puta, bins=np.arange(.5, 7.5))[0]
puno_bacanja = np.histogram(puno_puta, bins=np.arange(.5, 7.5))[0]

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(8,3))
ax1.bar(np.arange(1,7), malo_bacanja)
ax2.bar(np.arange(1,7), puno_bacanja)

### što je veći broj uzoraka vjerovatnost se približava teorijskoj vrijednosti događaja...
```

### 2. Jednostavni i složeni događaji (1/2)

```python
b = ss.distributions.binom
for bacanja in [5, 10, 20, 40, 80]:
  uspjeh = np.arange(bacanja)
  dist = b.pmf(uspjeh, bacanja, .5)
  plt.hist(uspjeh, bacanja, weights=dist)
plt.xlim(0,60)
```

```python
kolicina = [1,3,2]
cijena = [1.50, 1.44, 8.]
parcijalni_zbroj = []

for k, c in zip(kolicina, cijena):
  parcijalni_zbroj.append(k * c)

sum(parcijalni_zbroj)

```

### Alternative računanja

```python
kol1 = np.array([1,3,2])
cij1 = np.array([1.50, 1.44, 8.])

np.sum(kol1 * cij1)
```

### Funkcija "dot()" se najviše koristi

```python
print(kol1.dot(cij1), np.dot(kol1, cij1), kol1 @ cij1, sep='\n')
```

```python
dobitci = np.array([1., -.5])
vjerojatnost = np.array([.5, .5])
np.dot(dobitci, vjerojatnost)
```

```python
def je_paran(n):
  return n%  2 == 0
```

```python
dobitak2 = 0.0
for bacanje in range(10000):
  rez = np.random.randint(1,7)
  dobitak2 += 1.0 if je_paran(rez) else -0.5
print(dobitak2)
```
